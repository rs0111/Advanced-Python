```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_ice = pd.read_csv('../../data/icecream.csv')
```

```python
df_ice
```

```python
y = pd.Series(df_ice.expenditure_yen.values, index=pd.date_range('2003-1', periods=len(df_ice), freq='M'))
y.plot()
```

```python
y
```

```python
from statsmodels.tsa import stattools

ctt = stattools.adfuller(y[:100], regression="ctt")
ct = stattools.adfuller(y[:100], regression="ct")
c = stattools.adfuller(y[:100], regression="c")
n = stattools.adfuller(y[:100], regression="n")
print("ctt:")
print(ctt)
print("ct:")
print(ct)
print("c:")
print(c)
print("n:")
print(n)
```

```python
y_diff = y.diff()[:100].dropna()
y_diff.plot()
```

```python
ctt = stattools.adfuller(y_diff, regression="ctt")
ct = stattools.adfuller(y_diff, regression="ct")
c = stattools.adfuller(y_diff, regression="c")
n = stattools.adfuller(y_diff, regression="n")
print("ctt:")
print(ctt)
print("ct:")
print(ct)
print("c:")
print(c)
print("n:")
print(n)
```

```python
y_diff
```

```python
from statsmodels.tsa import ar_model
model = ar_model.AutoReg(y_diff[:80], lags=11)
```

```python
res = model.fit()
```

```python
forecast = res.forecast(30)
```

```python
fig = plt.figure(figsize=(15, 10))
ax = fig.add_subplot(2, 1, 1)
plt.plot(y_diff)
plt.plot(forecast)
plt.show()
```
